```python
# Open csv file
with open('Eng_Kin-Paralleldata.csv') as f:
    df = pd.read_csv(f)
df.head()
```

```python
# select columns to work on

df2 = df.loc[:, ["English", "Kinyarwanda"]]
```

```python
df2
```

```python
# Contractions for English words

df2["no_english_contract"] = df2["English"].apply(lambda x:[contractions.fix(word) for word in x.split()])
df2.head()

```

```python
 # Contraction for Kinyarwanda

df2["no_kinyarwanda_contract"] = df2["Kinyarwanda"].apply(lambda x:[contractions.fix(word) for word in x.split()])
df2.head()
```

```python

# Expanding the contraction to be tokenized seperately in english
# Remove whitespaces


df2['english_string'] = [' '.join(map(str, e)) for e in df2['no_english_contract']]
df2.head()
```

```python

# Expanding the contraction to be tokenized seperately in kinyarwanda
# Remove whitespaces


df2['kinyarwanda_string'] = [' '.join(map(str, e)) for e in df2['no_kinyarwanda_contract']]
df2.head()
```

```python
pretrained_model = "lid.176.bin" 
model = fasttext.load_model(pretrained_model)

languages = []
for sent in df2['english_string']:
    language = model.predict(sent)[0]
    languages.append(str(language)[11:13])
df2['English languages'] = languages
```
